```python
for i in range(data.shape[0]):
    if(data[i][0]<5.1):
        data[i][0]=1
    elif(data[i][0]<5.8):
        data[i][0]=2
    elif(data[i][0]<6.4):
        data[i][0]=3
    else:
        data[i][0]=4
        
    if(data[i][1]<2.8):
        data[i][1]=1
    elif(data[i][1]<3.0):
        data[i][1]=2
    elif(data[i][1]<3.3):
        data[i][1]=3
    else:
        data[i][1]=4
        
    if(data[i][2]<1.6):
        data[i][2]=1
    elif(data[i][2]<4.35):
        data[i][2]=2
    elif(data[i][2]<5.1):
        data[i][2]=3
    else:
        data[i][2]=4
        
    if(data[i][3]<0.3):
        data[i][3]=1
    elif(data[i][3]<1.3):
        data[i][3]=2
    elif(data[i][3]<1.8):
        data[i][3]=3
    else:
        data[i][3]=4
        
    
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(data,target)
```

```python
#TRAINING
newdict={}
outputs=np.unique(y_train)
for i in outputs:
    newdict[i]={}
for i in newdict:
    newdict[i]['Sepal Length']={}
    newdict[i]['Sepal Width']={}
    newdict[i]['Petal Length']={}
    newdict[i]['Petal Width']={}
#     newdict[i]['Total']=0
for i in newdict:
    for j in newdict[i]:
        newdict[i][j][1]=0
        newdict[i][j][2]=0
        newdict[i][j][3]=0
        newdict[i][j][4]=0
for i in newdict:
    newdict[i]['Total']=0
    
```

```python
newdict
```

```python
col=['Sepal Length','Sepal Width','Petal Length','Petal Width']
for i in range(x_train.shape[0]):
    for j in range(len(col)):
        newdict[y_train[i]][col[j]][x_train[i][j]]+=1
    newdict[y_train[i]]['Total']+=1
    
```

```python
newdict
```

```python
#TESTING
count1=0
for i in range(x_test.shape[0]):
    max1=-1*math.inf
    for j in range(len(outputs)):
        finalval=1
        for k in range(len(col)):
            temp=newdict[outputs[j]][col[k]][x_test[i][k]]/newdict[outputs[j]]['Total']
#             print(temp)
            finalval+=temp
#         print(i," ",j," ",finalval)
        finalval+=newdict[outputs[j]]['Total']/x_train.shape[0]
        if(finalval>max1):
            max1=finalval
            out=j
    print("Predicted: ",out," --  Actual ",y_test[i])
    if(out==y_test[i]):
        count1+=1
            
        
        
        
```
